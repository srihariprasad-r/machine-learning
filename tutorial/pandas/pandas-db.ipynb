```python
import numpy as np
import pandas as pd
```

# **Series**

```python
# based on numpy array
list_1 = ['a','b','c','d']
labels = [1,2,3,4]
df = pd.Series(data=list_1, index=labels)
df

arr_1 = np.array([1,2,3,4])
ser_1 = pd.Series(data=arr_1)
ser_1

dict_1 = {'apple': 1, 'banana': 2, 'orange': 3}
ser_2 = pd.Series(data=dict_1)
ser_2['apple']

# Series will operate on labels
arr_4 = np.array([4,5, 6, 7])
ser_2 = pd.Series(arr_4)
ser_1 + ser_2
```

# **Dataframes**

```python
arr_5 = np.random.randint(10, 15, size=(2, 3))
# takes arr, row labels and column labels
df_1 = pd.DataFrame(arr_5, ['A','B'], ['C','D','E'])
df_1

# if dict is passed and if it has scaler values, index should be passed
dict_2 = {'f_name': 'abc', 'l_name': 'ced', 'city': 'some'}
df_2 = pd.DataFrame(data=dict_2, index=[0])
df_2

# if dict is passed as collection, index can be excluded
dict_3 = {'col_1': pd.Series([1,2], index=['c0','c1']), 'col_2': pd.Series([1,2,4], index=['c0','c1','c2'])}
df_3 = pd.DataFrame(data=dict_3)
df_3

print(df_1)

# use '[[' for multiple columns to be queried
df_1[['D','E']]

# if you need to retreive location of row, query based on index
# if you need column alone, prefix ':' before column name
df_1.loc['A']

dict_4 = {'col_1': pd.Series([1,2],index=[20,80]), 'col_2': pd.Series([1,2,4], index=[20,10,1])}
df_4 = pd.DataFrame(data=dict_4)
print(df_4)

# iloc = index location(similar to python arrays); loc = label location
df_4.iloc[0]  # returns row at index 0
df_4.loc[0] # returns error - no index with value 0 exists
```
